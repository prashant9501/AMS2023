```python
f = lambda x: 2*x**2 - 3*x + 1
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.25, 1.25, 25)
y = 10*f(x)
```

```python
plt.plot(x,y, '-o')
plt.grid("on")
```

```python
from scipy.optimize import minimize, minimize_scalar

minimize(f, 0.8)
```

```python
minimize(f, 0.6, method='Nelder-Mead')
```

```python
minimize_scalar(f, method='brent')
```
